# Population Data from CSV

This notebooks reads sample population data from `data/atlantis.csv` and plots it using Matplotlib. Edit `data/atlantis.csv` and re-run this cell to see how the plots change!

```python
import matplotlib.pyplot as plt
import pandas

df = pandas.read_csv('../data/atlantisB.csv')
x = df['year']
y = df['deaths']
print(df)

plt.plot(x,y)
plt.title("Deaths of Atlantis")
plt.xlabel('Year')
plt.ylabel('Deaths')
plt.show()


dfYes = pandas.read_csv("../data/atlantisB.csv",usecols=["deaths"])

print("The dataframe is:")
print(dfYes)
```
